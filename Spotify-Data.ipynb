```python
import pandas as pd 
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt 
%matplotlib inline 
```

```python
sp_track=pd.read_csv(r"tracks.csv")
```

# Extracting Most Popular Album 

```python
most_popular = sp_track.query('popularity>90', inplace=False).sort_values('popularity',ascending=False)
most_popular[:10]
```

# Set release Date as Index 

```python
sp_track.set_index('release_date', inplace= True)
```

# Track the artist at location 18 

```python
sp_track[['artists']].iloc[18]
```

# make  Duration coloumn and Drop Duration_ms Coloumn 

```python
sp_track['duration'] = sp_track['duration_ms'].apply(lambda x : round(x/1000))
sp_track.drop('duration_ms',axis=1,inplace=True)
```

```python
sp_track.duration.head()
```

```python
sp_track.head()
```

# Find Correlation and drop of useless coloumns

```python
sp_corr = sp_track.drop(["key","mode","explicit","id","name",'artists','id_artists'] , axis= 1).corr(method="pearson")
sp_corr
```

# Plot a Heat map To Visualize The Given data 

```python
plt.figure(figsize=(12,8))
sns.heatmap(sp_corr,annot=True,fmt=".1g",linecolor="black",linewidths=1,cmap="inferno")
plt.title("Corelation Heatmap Between variables",weight="bold")
```

# Make a Sample data Frame to visualize Between Loudness and Energy 

```python
Sample_df = sp_track.sample(int(0.004*len(sp_track)))
Sample_df
```

# Plot a RegPlot to discover The Relation between Them

```python
plt.figure(figsize=(10,6))
sns.regplot(Sample_df,y="loudness",x="energy" , color="c")
plt.title("Loudness Vs Energy",weight = "bold")
```

# Plot The Same Between Popularity vs acousticness

```python
plt.figure(figsize=(10,6))
sns.regplot(Sample_df,y="popularity",x="acousticness" , color="b")
plt.title(" Popularity vs Acousticness",weight = "bold")
```

# calculations to Check No Of Songs Per Year 

```python
sp_track['dates'] = sp_track.index.get_level_values('release_date')
sp_track['dates']=pd.to_datetime(sp_track['dates'],format='mixed')
years = sp_track['dates'].dt.year
years
```

```python
# plt.figure(figsize=(18,6))
sns.displot(years ,kind="hist",discrete=True,aspect=2,height=5)
plt.title("No Of songs Per year",weight ='bold')
```

# Calculation On the Basis of Year Vs Duartion 

```python
total_dr = sp_track.duration
# total_dr
plt.figure(figsize=(18,8))
sns.barplot(x=years,y=total_dr,errwidth=False,palette='spring')
plt.title('Year vs Duration',weight = 'bold')
plt.xticks(rotation = 90)
```

```python
plt.figure(figsize=(10,5))
sns.set_style(style='whitegrid')
sns.lineplot(x=years,y=total_dr)
plt.title("Years Vs Duration ",weight = "bold")
```
